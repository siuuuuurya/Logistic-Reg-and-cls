```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import accuracy_score , classification_report 
```

```python
df = pd.read_csv("/Users/siuuuuurya/Desktop/excelfiles/login_data.csv")
```

```python
X = df.drop(['Suspicious'], axis = 1)
```

```python
Y = df.Suspicious
```

```python
x_train , x_test , y_train , y_test = train_test_split(X , Y , test_size= 0.20, random_state = 42) 
```

```python
Model = LogisticRegression()
Model
```

```python
Model.fit(x_train,y_train)
```

```python
predicted_values = Model.predict(x_test)
```

```python
predicted_values
```

```python
accuracy_score(y_test, predicted_values)
```

```python
Score = Model.score(x_test, y_test)
Score
```

```python
def Input_cred():
    logintime = int(entry_logintime.get())
    loginlocation = int(entry_loginlocation.get())
    devicetype = int(entry_devicetype.get())
    failedattempts = int(entry_failedattempts.get())

    features = [logintime, loginlocation, devicetype, failedattempts]

    prediction = Model.predict([features])

    if prediction[0] == 1:
        messagebox.showinfo("Result", "SUS login")
    else:
        messagebox.showinfo("Result", "NOT SO SUS login")
```

```python
root = tk.Tk()
root.title("Suspecious Login Detection")
root.geometry("800x800")
```

```python
root
```

```python
Lbl1 = tk.Label(root, text = "LoginTime(0 to 24)").pack()
entry_logintime = tk.Entry(root)
entry_logintime.pack()

Lbl2 = tk.Label(root, text = "entry_loginlocation(0 or 1)").pack()
entry_loginlocation= tk.Entry(root)
entry_loginlocation.pack()

Lbl3 = tk.Label(root, text = "devicetype(0 or 1").pack()
entry_devicetype = tk.Entry(root)
entry_devicetype.pack()

Lbl4 = tk.Label(root, text = "failedattempts").pack()
entry_failedattempts = tk.Entry(root)
entry_failedattempts.pack()

result_Lbl = tk.Label(root, text="")
result_Lbl.pack()

btn1 = tk.Button(root,
                 text="click here",
                 command=Input_cred)
btn1.pack(padx=90)
 
```
