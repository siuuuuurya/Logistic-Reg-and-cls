```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score, classification_report
#import seaborn as sns
```

```python
df = pd.read_csv("/Users/siuuuuurya/Desktop/excelfiles/admission_data.csv")
```

```python
df_x =  df.drop(['Admitted'], axis = 1)
```

```python
df_y = df.Admitted
```

```python
x_train , x_test ,y_train ,y_test = train_test_split(df_x , df_y, test_size=0.1 ,random_state= 42) 
```

```python
Model = LogisticRegression()
Model.fit(x_train, y_train)

predicted = Model.predict(x_test)
```

```python
predicted
```

```python
df.info()
```

```python
Model.score(x_test,y_test)
```

```python
print("Accuracy Score : ", accuracy_score(y_test,predicted))
```

```python
print("Classification Report : ", classification_report(y_test,predicted))
```
